```python
pge2_wc = WordCloud(
    background_color='skyblue',
    max_words=2000,
    stopwords=stopwords)

pge2_wc.generate(pge2)
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 


plt.imshow(pge2_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
